# **# prompt: Emotional reaction of Hindi and any text**

```python


from textblob import TextBlob #thats libary of Emotional reaction of Hindi and any text TextBlob

def get_emotion(text):
  blob = TextBlob(text)
  sentiment = blob.sentiment.polarity
  if sentiment > 0:
    return "positive"
  elif sentiment < 0:
    return "negative"
  else:
    return "neutral"

df["emotion"] = df["text"].apply(get_emotion)
df
```

```python
# prompt: count language number of all text

language_counts = {}
for text in df['text']:
    languages = detect_langs(text)
    for lang in languages:
        if lang.lang not in language_counts:
            language_counts[lang.lang] = 0
        language_counts[lang.lang] += 1

# Print the language counts
for language, count in language_counts.items():
    print(f"{language}: {count}")
```

```python
language_df = pd.DataFrame.from_dict(language_counts, orient='index', columns=['Count']).sort_values(by='Count', ascending=False)
language_df.index.name = 'Language'
language_df.columns = ['Count']
language_df
```

```python
import re
hashtag_pattern = r'#\w+'
hashtag_count = []

for text in df['text']:
    hashtags = re.findall(hashtag_pattern, text)
    for hashtag in hashtags:

        found = False
        for idx, item in enumerate(hashtag_count):
            if item[0] == hashtag:
                hashtag_count[idx] = (hashtag, item[1] + 1)
                found = True
                break
        if not found:
            hashtag_count.append((hashtag, 1))
for item in hashtag_count:
    print(item)
```

```python

# prompt: replace the all hashtags data frame how many hashtags same count simple code

hashtag_count_df = pd.DataFrame(hashtag_count, columns=['Hashtag', 'Count'])
hashtag_count_df = hashtag_count_df.sort_values(by='Count', ascending=False)
hashtag_count_df = hashtag_count_df.drop_duplicates(subset='Count', keep='first')
print(hashtag_count_df)
```
